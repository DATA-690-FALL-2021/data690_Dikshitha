```python
import pandas as pd
import plotly
print('plotly version',plotly.__version__)
import plotly.graph_objects as go
import plotly.express as px
import plotlywidget
```

```python
df = pd.read_csv('wdi_data.csv')
df.head()
```

```python
df_new= df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
df_new.tail(20)
```

```python
df_ind = df_new[df_new['Country Code']== 'IND']
df_ind
```

```python
df_ind_stocks = df_ind.query("indicator in(['CM.MKT.TRNR','CM.MKT.TRAD.GD.ZS','CM.MKT.TRAD.CD'])")
df_ind_stocks
```

```python
df_pivot = df_ind_stocks.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```python
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "CM.MKT.TRAD.CD",
    title = "Stocks",
    labels = {"Year":"Year","CM.MKT.TRAD.CD":"Stocks"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    
)

fig.show()
```
